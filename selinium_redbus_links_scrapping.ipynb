```python
#open browser
driver = webdriver.Edge()

driver.get('https://www.redbus.in/online-booking/tsrtc/?utm_source=rtchometile')

time.sleep(3)
driver.maximize_window()
```

```python

#retrive bus links and route

wait = WebDriverWait(driver,20)

def Telugna_link_route(path):
    LINKS_TELUGANA=[]
    ROUTE_TELUGNA=[]
    
    for i in range(1,4):
        paths=driver.find_elements(By.XPATH,path)
        
        #retrive the route links
        for links in paths:
            d = links.get_attribute("href")
            LINKS_TELUGANA.append(d)
            
        #retrive names of routes
        for route in paths:
            ROUTE_TELUGNA.append(route.text)
            
        try:
            #Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            time.sleep(3)
            next_button.click()
            
        except NoSuchElementException:
            print(f"No more pages in paginate at step {i} ")
            driver.quit()
            break
        
    return LINKS_TELUGANA,ROUTE_TELUGNA

LINKS_TELUGANA,ROUTE_TELUGNA = Telugna_link_route("//a[@class='route']")
```

```python
df_t = pd.DataFrame({"Route_name":ROUTE_TELUGNA , "Route_link":LINKS_TELUGANA})
df_t 
```

```python
df_t.to_csv("telegana.csv",index=False)
```

```python
#retrive bus links and route 

def link_route(path):
    wait = WebDriverWait(driver,20)

    LINKS=[]
    ROUTE=[]
    
    for i in range(1,7):
        paths=driver.find_elements(By.XPATH,path)
        
        #retrive the route links
        for links in paths:
            d = links.get_attribute("href")
            LINKS.append(d)
            
        #retrive names of routes
        for route in paths:
            ROUTE.append(route.text)
            
        try:
            #Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            time.sleep(3)
            next_button.click()
            
        except NoSuchElementException:
            print(f"No more pages in paginate at step {i} ")
            driver.quit()
            break
        
    return LINKS,ROUTE
```

```python
#ksrtc-kerala
#open browser
driver = webdriver.Edge()

driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")

time.sleep(3)
driver.maximize_window()
```

```python
LINKS,ROUTE = link_route("//a[@class='route']")
df_k = pd.DataFrame({"Route_name": ROUTE , "Route_link":LINKS})
df_k 
```
